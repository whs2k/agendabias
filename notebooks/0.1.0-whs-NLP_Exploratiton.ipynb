```python
from newspaper import Article
import spacy
from spacy import displacy
#from spacy import TextCategorizer

#import en_core_web_sm
#nlp = en_core_web_sm.load()
nlp = spacy.load("en_core_web_sm")

import nltk
nltk.download('punkt')
```

```python
url = 'http://fox13now.com/2013/12/30/new-year-new-laws-obamacare-pot-guns-and-drones/'
url2 = 'https://slate.com/news-and-politics/2020/11/georgia-secretary-of-state-brad-raffensperger-lindsey-graham-signature-mismatch-ballots.html'
url3 = 'https://www.cnn.com/2020/11/16/politics/georgia-secretary-of-state-lindsey-graham-ballots-cnntv/index.html'
```

```python
article = Article(url3)
article.download()
article.html
article.parse()
article.text
article.top_image
heuristics = ['obviously lying','totally false']
heuristics_spacy = [nlp(heur) for heur in heuristics]
doc = nlp(article.text)
for doc_heur in heuristics_spacy:
    for sent in doc.sents:
        sim_score = doc_heur.similarity(sent)
        if sim_score > 0.4:
            print(doc_heur.text)
            print(sim_score)
            print(sent.text)
            print()
```
